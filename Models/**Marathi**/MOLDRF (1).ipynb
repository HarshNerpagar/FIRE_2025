```python
import pandas as pd
import numpy as np
import re
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import classification_report, f1_score
from sklearn.preprocessing import LabelEncoder
```

```python
# Using your paths
train_df = pd.read_excel('/content/MOLDV2_train.csv.xlsx')
test_df = pd.read_excel('/content/MOLDV2_test.csv.xlsx')
```

```python
def clean_text(text):
    text = str(text)
    text = re.sub(r"http\S+|@\w+|#\w+|\d+|[^\w\s]", "", text)
    return re.sub(r"\s+", " ", text).strip()
```

```python
tfidf = TfidfVectorizer(max_features=5000, ngram_range=(1, 2))
```

```python

print("\n--- Subtask A ---")
train_a = train_df.dropna(subset=['tweet', 'subtask_a']).copy()
test_a = test_df.dropna(subset=['tweet', 'subtask_a']).copy()

train_a['clean_tweet'] = train_a['tweet'].apply(clean_text)
test_a['clean_tweet'] = test_a['tweet'].apply(clean_text)

label_a = LabelEncoder()
y_train_a = label_a.fit_transform(train_a['subtask_a'])
y_test_a = label_a.transform(test_a['subtask_a'])

X_train_a = tfidf.fit_transform(train_a['clean_tweet'])
X_test_a = tfidf.transform(test_a['clean_tweet'])

rf_a = RandomForestClassifier(n_estimators=100, random_state=42)
rf_a.fit(X_train_a, y_train_a)

preds_a = rf_a.predict(X_test_a)
print(classification_report(y_test_a, preds_a, target_names=label_a.classes_))
print("Macro F1 (Subtask A):", f1_score(y_test_a, preds_a, average='macro'))
```

```python
print("\n--- Subtask B ---")
train_b = train_df.dropna(subset=['tweet', 'subtask_b']).copy()
test_b = test_df.dropna(subset=['tweet', 'subtask_b']).copy()

train_b['clean_tweet'] = train_b['tweet'].apply(clean_text)
test_b['clean_tweet'] = test_b['tweet'].apply(clean_text)

label_b = LabelEncoder()
y_train_b = label_b.fit_transform(train_b['subtask_b'])
y_test_b = label_b.transform(test_b['subtask_b'])

X_train_b = tfidf.transform(train_b['clean_tweet'])
X_test_b = tfidf.transform(test_b['clean_tweet'])

rf_b = RandomForestClassifier(n_estimators=100, random_state=42)
rf_b.fit(X_train_b, y_train_b)

preds_b = rf_b.predict(X_test_b)
print(classification_report(y_test_b, preds_b, target_names=label_b.classes_))
print("Macro F1 (Subtask B):", f1_score(y_test_b, preds_b, average='macro'))
```

```python
print("\n--- Subtask C ---")
train_c = train_df.dropna(subset=['tweet', 'subtask_c']).copy()
test_c = test_df.dropna(subset=['tweet', 'subtask_c']).copy()

train_c['clean_tweet'] = train_c['tweet'].apply(clean_text)
test_c['clean_tweet'] = test_c['tweet'].apply(clean_text)

label_c = LabelEncoder()
y_train_c = label_c.fit_transform(train_c['subtask_c'])
y_test_c = label_c.transform(test_c['subtask_c'])

X_train_c = tfidf.transform(train_c['clean_tweet'])
X_test_c = tfidf.transform(test_c['clean_tweet'])

rf_c = RandomForestClassifier(n_estimators=100, random_state=42)
rf_c.fit(X_train_c, y_train_c)

preds_c = rf_c.predict(X_test_c)
print(classification_report(y_test_c, preds_c, target_names=label_c.classes_))
print("Macro F1 (Subtask C):", f1_score(y_test_c, preds_c, average='macro'))
```

```python
!pip install -q scikit-learn openpyxl

import pandas as pd
import re
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import FeatureUnion
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, f1_score
```

```python
train_df = pd.read_excel('/content/MOLDV2_train.csv.xlsx')
test_df = pd.read_excel('/content/MOLDV2_test.csv.xlsx')
```

```python
def clean_marathi_text(text):
    text = str(text)
    text = re.sub(r"http\S+|@\w+|#\w+|\d+", "", text)  # remove links, hashtags, digits
    text = re.sub(r"[^\w\s\u0900-\u097F]", "", text)  # remove symbols but keep Devanagari
    return re.sub(r"\s+", " ", text).strip()
```

```python
print("\n--- Subtask A ---")
train_a = train_df.dropna(subset=['tweet', 'subtask_a']).copy()
test_a = test_df.dropna(subset=['tweet', 'subtask_a']).copy()

train_a['clean_tweet'] = train_a['tweet'].apply(clean_marathi_text)
test_a['clean_tweet'] = test_a['tweet'].apply(clean_marathi_text)

label_a = LabelEncoder()
y_train_a = label_a.fit_transform(train_a['subtask_a'])
y_test_a = label_a.transform(test_a['subtask_a'])
```

```python
# Combine word and character TF-IDF
word_vectorizer = TfidfVectorizer(analyzer='word', ngram_range=(1, 2), max_features=3000)
char_vectorizer = TfidfVectorizer(analyzer='char_wb', ngram_range=(3, 5), max_features=3000)

combined_vectorizer = FeatureUnion([('word', word_vectorizer), ('char', char_vectorizer)])

X_train_a = combined_vectorizer.fit_transform(train_a['clean_tweet'])
X_test_a = combined_vectorizer.transform(test_a['clean_tweet'])
```

```python
param_grid = {
    'n_estimators': [100, 300],
    'max_depth': [10, 30, None],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'max_features': ['sqrt', 'log2']
}

grid_rf = GridSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_grid=param_grid,
    scoring='f1_macro',
    cv=3,
    n_jobs=-1,
    verbose=1
)

grid_rf.fit(X_train_a, y_train_a)
```

```python
best_rf = grid_rf.best_estimator_
preds_a = best_rf.predict(X_test_a)

print("\n Best Parameters:", grid_rf.best_params_)
print(" Classification Report – Subtask A:")
print(classification_report(y_test_a, preds_a, target_names=label_a.classes_))

macro_f1 = f1_score(y_test_a, preds_a, average='macro')
print(f" Macro F1 (Tuned RF): {macro_f1:.4f}")
```

```python
test_a['predicted'] = label_a.inverse_transform(preds_a)
test_a[['tweet', 'subtask_a', 'predicted']].to_csv("rf_subtask_a_predictions.csv", index=False)
```

```python
print("\n--- Subtask B ---")
train_b = train_df.dropna(subset=['tweet', 'subtask_b']).copy()
test_b = test_df.dropna(subset=['tweet', 'subtask_b']).copy()

train_b['clean_tweet'] = train_b['tweet'].apply(clean_marathi_text)
test_b['clean_tweet'] = test_b['tweet'].apply(clean_marathi_text)

label_b = LabelEncoder()
y_train_b = label_b.fit_transform(train_b['subtask_b'])
y_test_b = label_b.transform(test_b['subtask_b'])

X_train_b = combined_vectorizer.transform(train_b['clean_tweet'])
X_test_b = combined_vectorizer.transform(test_b['clean_tweet'])
```

```python
rf_b_tuned = RandomForestClassifier(
    n_estimators=300,
    max_depth=None,
    max_features='sqrt',
    min_samples_split=5,
    min_samples_leaf=1,
    random_state=42

)

rf_b_tuned.fit(X_train_b, y_train_b)
preds_b = rf_b_tuned.predict(X_test_b)

print(" Classification Report – Subtask B:")
print(classification_report(y_test_b, preds_b, target_names=label_b.classes_))
f1_b = f1_score(y_test_b, preds_b, average='macro')
print(f" Macro F1 (Subtask B): {f1_b:.4f}")
```

```python
print("\n--- Subtask C ---")
train_c = train_df.dropna(subset=['tweet', 'subtask_c']).copy()
test_c = test_df.dropna(subset=['tweet', 'subtask_c']).copy()

train_c['clean_tweet'] = train_c['tweet'].apply(clean_marathi_text)
test_c['clean_tweet'] = test_c['tweet'].apply(clean_marathi_text)

label_c = LabelEncoder()
y_train_c = label_c.fit_transform(train_c['subtask_c'])
y_test_c = label_c.transform(test_c['subtask_c'])

X_train_c = combined_vectorizer.transform(train_c['clean_tweet'])
X_test_c = combined_vectorizer.transform(test_c['clean_tweet'])
```

```python
rf_c_tuned = RandomForestClassifier(
    n_estimators=300,
    max_depth=None,
    max_features='sqrt',
    min_samples_split=5,
    min_samples_leaf=1,
    random_state=42
)

rf_c_tuned.fit(X_train_c, y_train_c)
preds_c = rf_c_tuned.predict(X_test_c)

print(" Classification Report – Subtask C:")
print(classification_report(y_test_c, preds_c, target_names=label_c.classes_))
f1_c = f1_score(y_test_c, preds_c, average='macro')
print(f" Macro F1 (Subtask C): {f1_c:.4f}")
```

```python
test_b['predicted_b'] = label_b.inverse_transform(preds_b)
test_c['predicted_c'] = label_c.inverse_transform(preds_c)

test_b[['tweet', 'subtask_b', 'predicted_b']].to_csv("rf_subtask_b_predictions.csv", index=False)
test_c[['tweet', 'subtask_c', 'predicted_c']].to_csv("rf_subtask_c_predictions.csv", index=False)
```

```python

from sklearn.pipeline import Pipeline

# Subtask A
pipeline_a_tuned = Pipeline([
    ('vectorizer', combined_vectorizer),
    ('clf', best_rf)
])
joblib.dump(pipeline_a_tuned, '/content/drive/MyDrive/tuned_random_forest_pipeline_subtask_a.pkl')

# Subtask B
pipeline_b_tuned = Pipeline([
    ('vectorizer', combined_vectorizer),
    ('clf', rf_b_tuned)
])
joblib.dump(pipeline_b_tuned, '/content/drive/MyDrive/tuned_random_forest_pipeline_subtask_b.pkl')

# Subtask C
pipeline_c_tuned = Pipeline([
    ('vectorizer', combined_vectorizer),
    ('clf', rf_c_tuned)
])
joblib.dump(pipeline_c_tuned, '/content/drive/MyDrive/tuned_random_forest_pipeline_subtask_c.pkl')

print(" Pipelines saved to Google Drive")
```
