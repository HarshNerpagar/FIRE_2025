```python
import joblib
import pandas as pd
from sklearn.metrics import classification_report

# ========= 1. Load Saved Pipelines =========
pipeline_a = joblib.load('/content/drive/MyDrive/tuned_random_forest_pipeline_subtask_a.pkl')
pipeline_b = joblib.load('/content/drive/MyDrive/tuned_random_forest_pipeline_subtask_b.pkl')
pipeline_c = joblib.load('/content/drive/MyDrive/tuned_random_forest_pipeline_subtask_c.pkl')

# ========= 2. Load Hindi Datasets =========
hindi_train = pd.read_excel("/content/hasoc_2020_hi_train.xlsx")

try:
    hindi_test = pd.read_csv("/content/hindi_test_1509.csv", encoding="latin-1")
except UnicodeDecodeError:
    try:
        hindi_test = pd.read_csv("/content/hindi_test_1509.csv", encoding="iso-8859-1")
    except UnicodeDecodeError:
        print("Could not decode Hindi test file, loading empty DataFrame.")
        hindi_test = pd.DataFrame()

# ========= 2a. Ensure 'cleaned' column exists =========
for df in [hindi_train, hindi_test]:
    if not df.empty:
        df.rename(columns={"text": "cleaned"}, inplace=True)

# ========= 3. Map Labels to Match Subtasks =========
binary_map = {'HOF': 1, 'NOT': 0}
multi_map = {'OFFN': 0, 'HATE': 1, 'PRFN': 2}  # adjust if needed

for df in [hindi_train, hindi_test]:
    if not df.empty:
        df['Label_A'] = df['task1'].map(binary_map)  # Binary labels
        df['Label_B'] = df['task2'].map(multi_map)  # Multi-class labels (Subtask B)
        df['Label_C'] = df['task2'].map(multi_map)  # Reuse for Subtask C

# ========= 4. Define Evaluation Function =========
def evaluate_dataset(name, df, pipeline, label_col):
    if df.empty:
        print(f"\nSkipping {name} (empty dataset).")
        return None
    print(f"\n=== {name} ===")
    y_true = df[label_col].dropna()
    y_pred = pipeline.predict(df.loc[y_true.index, "cleaned"])
    print(classification_report(y_true, y_pred))
    return y_pred

# ========= 5. Run Evaluations =========

## Subtask A: Binary
print("\n### Subtask A: Binary Classification ###")
evaluate_dataset("Train", hindi_train.rename(columns={'Label_A': 'Label'}), pipeline_a, "Label")
evaluate_dataset("Test", hindi_test.rename(columns={'Label_A': 'Label'}), pipeline_a, "Label")

## Subtask B: Multi-class
print("\n### Subtask B: Multi-class Classification ###")
evaluate_dataset("Train", hindi_train.rename(columns={'Label_B': 'Label'}), pipeline_b, "Label")
evaluate_dataset("Test", hindi_test.rename(columns={'Label_B': 'Label'}), pipeline_b, "Label")

## Subtask C: Multi-class
print("\n### Subtask C: Multi-class Classification ###")
evaluate_dataset("Train", hindi_train.rename(columns={'Label_C': 'Label'}), pipeline_c, "Label")
evaluate_dataset("Test", hindi_test.rename(columns={'Label_C': 'Label'}), pipeline_c, "Label")
```
